```python
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import pyplot as pl

import scipy.io
```

```python
tensor_arr_a = scipy.io.loadmat('test_glrsp_01.mat')
arr_a = tensor_arr_a['tensor_iht_arr']
m = list(tensor_arr_a['m_vector'][0])
mediana = np.median(arr_a, 1)
p25a = np.percentile(arr_a, 25, 1)
p75a = np.percentile(arr_a, 75, 1)

tensor_arr_b = scipy.io.loadmat('test_rauhut_01.mat')
arr_b = tensor_arr_b['tensor_iht_arr']
medianb = np.median(arr_b, 1)
p25b = np.percentile(arr_b, 25, 1)
p75b = np.percentile(arr_b, 75, 1)

tensor_arr_c = scipy.io.loadmat('test_cp_01.mat')
arr_c = tensor_arr_c['tensor_iht_arr']
medianc = np.median(arr_c, 1)
p25c = np.percentile(arr_c, 25, 1)
p75c = np.percentile(arr_c, 75, 1)
```

```python
idx = 2
print("Number of samples = ", m[idx])

# get data
arr_tpgd = tensor_arr_a['tensor_iht_arr'][idx]
arr_pgd_tucker = tensor_arr_b['tensor_iht_arr'][idx]
arr_pgd_cp = tensor_arr_c['tensor_iht_arr'][idx]

# make data frames
df_tpgd = pd.DataFrame(data = arr_tpgd, columns=['Normalized Estimation Error'])
df_tpgd['Method'] = 'TPGD'

df_pgd_tucker = pd.DataFrame(data = arr_pgd_tucker, columns=['Normalized Estimation Error'])
df_pgd_tucker['Method'] = 'PGD-Tucker'

df_pgd_cp = pd.DataFrame(data = arr_pgd_cp, columns=['Normalized Estimation Error'])
df_pgd_cp['Method'] = 'PGD-CP'

# concatemnate data frames
df_all = pd.concat([df_tpgd, df_pgd_tucker, df_pgd_cp])
df_all.head()

ax = sns.violinplot(y = 'Method', x = 'Normalized Estimation Error', data = df_all, inner = "point")
#ax.set_yscale("log")
#ax.set_ylim(-0.3, 2.5)
#ax.ylim([75, 1125, 1e-3, 10])

```
